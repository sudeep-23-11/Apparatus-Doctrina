```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("./Ukraine.csv")
df
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna(how="any", axis=0)
```

```python
print(df.groupby(["Searh"])["url"].count())
```

```python
plt.pie(df.groupby(["Searh"])["url"].count(), colors=['r', 'b', 'g', 'm', 'c'], radius=1, startangle=90, shadow=True,
        explode=[0.1, 0, 0, 0, 0], autopct="%1.1f%%", labels=["StandWithUkraine", "Ukraine NATO", "Ukraine border",
                                                              "Ukraine troops", "Ukraine war"])
plt.show()
```

```python
plt.hist(df["replyCount"], bins=10, range=[0, 5], rwidth=0.8, bottom=0, align="mid", linewidth=5, color='r',
         histtype="bar", orientation="vertical")
plt.xlabel("replyCount", color='r')
plt.ylabel("Frequency", color='r')
plt.show()
```

```python
plt.hist(df["retweetCount"], bins=10, range=[0, 5], rwidth=0.8, bottom=0, align="mid", linewidth=5, color='b',
         histtype="bar", orientation="vertical")
plt.xlabel("retweetCount", color='b')
plt.ylabel("Frequency", color='b')
plt.show()
```

```python
plt.hist(df["likeCount"], bins=10, range=[0, 5], rwidth=0.8, bottom=0, align="mid", linewidth=5, color='g',
         histtype="bar", orientation="vertical")
plt.xlabel("likeCount", color='g')
plt.ylabel("Frequency", color='g')
plt.show()
```

```python
plt.hist(df["quoteCount"], bins=10, range=[0, 5], rwidth=0.8, bottom=0, align="mid", linewidth=5, color='y',
         histtype="bar", orientation="vertical")
plt.xlabel("quoteCount", color='y')
plt.ylabel("Frequency", color='y')
plt.show()
```

```python
data={"lang":np.sort(df["lang"].unique()), "count":df.groupby(["lang"])["url"].count().to_numpy()}
_df=pd.DataFrame(data)
_df=_df[_df["count"]>=1000]
```

```python
plt.bar(_df["lang"], _df["count"], width=0.8, bottom=0, align="center", linewidth=5, color='m')
plt.xlabel("lang", color='m')
plt.ylabel("Frequency", color='m')
plt.show()
```

```python
data={"sourceLabel":np.sort(df["sourceLabel"].unique()), "count":df.groupby(["sourceLabel"])["url"].count().to_numpy()}
_df=pd.DataFrame(data)
_df=_df[_df["count"]>=1000]
```

```python
plt.bar(_df["sourceLabel"], _df["count"], width=0.8, bottom=0, align="center", linewidth=5, color='c')
plt.xlabel("sourceLabel", color='c')
plt.ylabel("Frequency", color='c')
plt.xticks(rotation=90)
plt.show()
```
