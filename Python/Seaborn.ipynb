```python
#Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
df
```

```python
sns.relplot(x="total_bill", y="tip", hue="time", row="sex", style="smoker", size="size",
            data=df, kind="scatter")
sns.relplot(x="size", y="tip", hue="smoker", col="sex", style="smoker", size="smoker",
            data=df, kind="line")
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="time", data=df, kind="strip").set(
    xlabel="day", ylabel="total_bill", title="tips")
sns.catplot(x="day", y="total_bill", hue="time", row="sex", data=df, kind="swarm")
sns.catplot(x="day", y="total_bill", hue="time", col="sex", data=df, kind="box")
sns.catplot(x="day", y="total_bill", hue="time", data=df, kind="violin")
sns.catplot(x="day", y="total_bill", hue="time", data=df, kind="boxen")
sns.catplot(x="day", y="total_bill", hue="time", data=df, kind="point")
sns.catplot(x="day", y="total_bill", hue="time", data=df, kind="bar")
sns.catplot(x="day", hue="time", data=df, kind="count")
plt.show()
```

```python
a=sns.FacetGrid(df, row="sex", col="time", hue="smoker")
a.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
a.add_legend()
plt.show()
```

```python
df=sns.load_dataset("penguins")
df
```

```python
sns.displot(x="flipper_length_mm", hue="species", row="sex", data=df, kind="hist")
sns.displot(x="flipper_length_mm", hue="species", col="sex", data=df, kind="kde")
sns.displot(x="flipper_length_mm", hue="species", rug=True, data=df, kind="ecdf")
sns.displot(x="flipper_length_mm", y="bill_length_mm", hue="species", data=df, kind="hist")
sns.displot(x="flipper_length_mm", y="bill_length_mm", hue="species", data=df, kind="kde")
plt.show()
```

```python
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", hue="species", data=df, kind="scatter")
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", hue="species", data=df, kind="kde")
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", hue="species", data=df, kind="hist")
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", data=df, kind="hex")
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", data=df, kind="reg")
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", data=df, kind="resid")
plt.show()
```

```python
b=sns.PairGrid(df, hue="species")
b.map_diag(sns.histplot)
b.map_offdiag(sns.scatterplot)
b.add_legend()
plt.show()
```

```python
c=sns.PairGrid(df, hue="species")
c.map_diag(sns.histplot)
c.map_upper(sns.scatterplot)
c.map_lower(sns.scatterplot)
c.add_legend()
plt.show()
```
