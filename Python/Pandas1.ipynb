```python
#Pandas1
import pandas as pd
```

```python
s=pd.Series([1, 2, 3, 4, 5])
print(s)
print(s.size, s.shape, s.dtype)
print(s.index, s.values, sep="\n")
```

```python
dict={0:["Sudeep", "Aryan", "Shivam", "Jatin", "Anshuman"], 1:[1, 2, 3, 4, 5], 2:[1000.0, 500.0, 200.0, 100.0, 40.0]}
df=pd.DataFrame(dict)
df.index=["First", "Second", "Third", "Fourth", "Fifth"]
df.columns=["Name", "Rank", "Price"]
df
```

```python
print(df.size, df.shape)
print(df.dtypes, df.index, df.columns, sep="\n")
print(df.head(3), df.tail(3), df.T, sep="\n")
```

```python
s=df["Price"]
print(s.max(), s.min(), s.mean(), s.median(), s.var(), s.std(), s.count())
print(s.value_counts(), s.unique(), s.tolist(), s.to_numpy(), s.describe(), sep="\n")
```

```python
df.iloc[2, 0]="Yash"
print(df.iat[2, 0])
df.loc["Third", "Name"]="Shivam"
print(df.at["Third", "Name"])
```

```python
print(df.loc[["Second", "Fourth"], ["Name", "Price"]])
print(df.loc[:, ["Rank"]], df.loc[["Third"], :], sep="\n")
print(df[(df["Rank"]<5) & (df["Price"]<1000)])          # |, ~
print(df.sort_index(axis=0, ascending=True), df.sort_values(by="Name", ascending=True), sep="\n")
```

```python
print(df.isnull(), df.notnull(), sep="\n")
print(df.fillna(value=0), df.dropna(how="any", axis=0),  sep="\n")         # all
```

```python
print(df.drop(["Second", "Fourth"], axis=0))
print(df.drop_duplicates(subset=["Name"], keep="first"))        # last, False
df1=df.copy()
df1.drop(["Third"], axis=0, inplace=True)
df1.reset_index(drop=True, inplace=True)
df1
```
