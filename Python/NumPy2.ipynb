```python
#NumPy2
import numpy as np
```

```python
A=np.arange(1, 4, 0.5)
print(np.ceil(A), np.floor(A), np.round(A), sep="\n")
print(np.square(A), np.power(A, 3), np.sqrt(A), np.cbrt(A), sep="\n")
print(np.log(A), np.log10(A), np.exp(A), sep="\n")
print(np.sin(A), np.cos(A), np.tan(A), sep="\n")
```

```python
A=np.array([2, 4, 1, 5, 3, 6])
print(np.sum(A), np.max(A), np.argmax(A), np.min(A), np.argmin(A))
print(np.mean(A), np.median(A), np.var(A), np.std(A))
print(np.sort(A), np.argsort(A), sep="\n")
print(np.delete(A, 3), np.insert(A, 2, 0), np.append(A, 0), sep="\n")
```

```python
A=np.array([[2, 4, 1], [5, 3, 6]])
print(np.sum(A, axis=0), np.sum(A, axis=1))
```

```python
print(np.zeros((2, 2)), np.ones((2, 2)), np.empty((2, 2)), sep="\n")
print(np.zeros_like(A), np.ones_like(A), np.empty_like(A), sep="\n")
print(np.eye(2, 2, k=1), np.identity(2), np.diag([1, 2]), sep="\n")
print(np.random.rand(2, 2), np.random.randint(5, size=(2, 2)), sep="\n")
```

```python
print(A.flatten(order='C'))         # F, A
print(np.transpose(A), np.swapaxes(A, 0, 1), sep="\n")
print(np.reshape(A, (6, 1), order='C'), np.resize(A, (9, 1)), sep="\n")
```

```python
A=np.array([[1, 2], [7, 8]])
B=np.array([[5, 6], [3, 4]])
print(np.where(A>5, A, -1), np.dot(A, B), sep="\n")
print(np.concatenate((A, B), axis=0), np.hstack([A, B]), np.vstack([A, B]), sep="\n")
```
