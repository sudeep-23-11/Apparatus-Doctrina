```python
import pandas as pd
```

```python
df=pd.read_csv('movie_ratings.csv')
df
```

```python
movie_mat=df.pivot_table(values='rating', index='user_id', columns='title')
movie_mat
```

```python
rating_count=df.groupby('title')['rating'].count()
rating_avg=df.groupby('title')['rating'].mean()
rating=pd.merge(rating_avg, rating_count, on='title')
rating.columns=['average', 'number']
rating
```

```python
def recommend_movies(movie) :
  similar_movie=movie_mat.corrwith(movie_mat[movie])
  corr=pd.DataFrame(similar_movie, columns=['Correlation'])
  corr=corr.join(rating['number'])
  corr.dropna(how='any', axis=0, inplace=True)
  print(corr[corr['number']>=100].sort_values(by='Correlation', ascending=False).head())
```

```python
recommend_movies('Star Wars (1977)')
```

```python
recommend_movies('Toy Story (1995)')
```

```python
recommend_movies('Shawshank Redemption, The (1994)')
```
