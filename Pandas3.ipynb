```python
#Pandas3
import pandas as pd
```

```python
dict1={"Name":["Sudeep", "Aryan", "Shivam"], "Rank":[1, 2, 3], "PriceMoney":[1000.0, 500.0, 200.0]}
df1=pd.DataFrame(dict1)
dict2={"Name":["Jatin", "Anshuman"], "Rank":[4, 5], "PriceMoney":[100.0, 40.0]}
df2=pd.DataFrame(dict2)
df3=pd.concat([df1, df2], ignore_index=True)
df3
```

```python
dict1={"Name":["Sudeep", "Aryan", "Shivam", "Jatin", "Anshuman"], "Rank":[1, 2, 3, 4, 5]}
df1=pd.DataFrame(dict1)
dict2={"PriceMoney":[1000.0, 500.0, 200.0, 100.0, 40.0]}
df2=pd.DataFrame(dict2)
df3=pd.concat([df1, df2], axis=1)
df3
```

```python
dict1={"Name":["Sudeep", "Aryan", "Shivam", "Jatin", "Anshuman"], "Rank":[1, 2, 3, 4, 5]}
df1=pd.DataFrame(dict1)
dict2={"Name":["Sudeep", "Aryan", "Shivam", "Jatin", "Anshuman"], "PriceMoney":[1000.0, 500.0, 200.0, 100.0, 40.0]}
df2=pd.DataFrame(dict2)
df3=pd.merge(df1, df2, on="Name")
df3
```
