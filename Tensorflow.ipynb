```python
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()
```

```python
a=tf.placeholder(tf.float32)
b=tf.placeholder(tf.float32)
c=tf.add(a, b)
d=tf.subtract(a, b)
e=tf.multiply(a, b)
f=tf.divide(a, b)
g=tf.mod(a, b)
sess=tf.Session()
print(sess.run(c, {a:[4, 6], b:[2, 3]}), end='\n\n')
print(sess.run(d, {a:[4, 6], b:[2, 3]}), end='\n\n')
print(sess.run(e, {a:[4, 6], b:[2, 3]}), end='\n\n')
print(sess.run(f, {a:[4, 6], b:[2, 3]}), end='\n\n')
print(sess.run(g, {a:[4, 6], b:[2, 3]}), end='\n\n')
```

```python
a=tf.constant(5, tf.float32)
b=tf.constant(5, tf.float32)
c=[tf.equal(a, b)]
d=[tf.greater(a, b), tf.greater_equal(a, b)]
e=[tf.less(a, b), tf.less_equal(a, b)]
sess=tf.Session()
print(sess.run(c), end='\n\n')
print(sess.run(d), end='\n\n')
print(sess.run(e), end='\n\n')
```

```python
a=tf.constant(5.5, tf.float32)
b=[tf.sin(a), tf.cos(a), tf.tan(a)]
c=[tf.log(a), tf.log1p(a), tf.exp(a)]
d=[tf.square(a), tf.sqrt(a), tf.pow(a, 5)]
e=[tf.ceil(a), tf.floor(a), tf.round(a)]
sess=tf.Session()
print(sess.run(b), end='\n\n')
print(sess.run(c), end='\n\n')
print(sess.run(d), end='\n\n')
print(sess.run(e), end='\n\n')
```

```python
w=tf.Variable([0.5], tf.float32)
b=tf.Variable([-0.5], tf.float32)
x=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
y_predict=(w*x)+b
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
print(sess.run(y_predict, {x:[1, 2, 3, 4, 5]}))
```

```python
delta=tf.subtract(y, y_predict)
loss=tf.reduce_sum(tf.square(delta))
print(sess.run(loss, {x:[1, 2, 3, 4, 5], y:[0, -1, -2, -3, -4]}))
```

```python
optimizer=tf.train.GradientDescentOptimizer(0.01)
train=optimizer.minimize(loss)
for i in range(0, 1000, 1) :
  sess.run(train, {x:[1, 2, 3, 4, 5], y:[0, -1, -2, -3, -4]})
print(sess.run([w, b]))
```
