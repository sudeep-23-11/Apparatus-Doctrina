```python
import tensorflow.compat.v1 as tf
import numpy as np
tf.disable_eager_execution()
```

```python
def linear(x) :
  return x
```

```python
def step(x) :
  y=np.greater(x, 5)
  z=y.astype('int32')
  return z
```

```python
def sigmoid(x) :
  y=np.exp(-x)
  z=1/(1+y)
  return z
```

```python
def tanh(x) :
  y=np.tanh(x)
  return y
```

```python
def relu(x) :
  y=[]
  for i in x :
    if i>=0 :
      y.append(i)
    else :
      y.append(0)
  return y
```

```python
def softmax(x) :
  y=np.exp(x)
  z=y/np.sum(y)
  return z
```

```python
A=np.array([7, -8, 9, -4, 5, -6, 1, -2, 3, 0])
print(linear(A), step(A), sigmoid(A), sep='\n\n', end='\n\n')
print(tanh(A), relu(A), softmax(A), sep='\n\n', end='\n\n')
```

```python
x=tf.placeholder(tf.float32)
sess=tf.Session()
print(sess.run(tf.nn.sigmoid(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.tanh(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.relu(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.softmax(x), feed_dict={x:A}), end='\n\n')
sess.close()
```
