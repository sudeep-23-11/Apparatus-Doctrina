```python
import math
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
def binary_step(x) :
    y = []
    for i in x :
        if i >= 0 :
            y.append(1)
        else :
            y.append(0)
    return np.array(y)
```

```python
def linear(x, n) :
    return x * n
```

```python
def sigmoid(x) :
    y = np.exp(-x)
    return 1 / (1 + y)
```

```python
def tanh(x) :
    y = np.exp(-x)
    z = np.exp(x)
    return (z - y) / (z + y)
```

```python
def reLU(x) :
    y = []
    for i in x :
        if i > 0 :
            y.append(i)
        else :
            y.append(0)
    return np.array(y)
```

```python
def leaky_reLU(x) :
    y = []
    for i in x :
        if i > 0 :
            y.append(i)
        else :
            y.append(i * 0.05)
    return np.array(y)
```

```python
def parameterised_reLU(x, n) :
    y = []
    for i in x :
        if i > 0 :
            y.append(i)
        else :
            y.append(i * n)
    return np.array(y)
```

```python
def ELU(x) :
    y = []
    for i in x :
        if i > 0 :
            y.append(i)
        else :
            y.append(math.exp(i) - 1)
    return np.array(y)
```

```python
def swish(x) :
    y = np.exp(-x)
    return x / (1 + y)
```

```python
def softmax(x) :
    y = np.exp(x)
    return y / np.sum(y)
```

```python
A = np.array([-8, -6, -4, -2, 0, 1, 3, 5, 7, 9])
print(binary_step(A), linear(A, 5), sigmoid(A), tanh(A), sep="\n\n", end="\n\n")
print(reLU(A), leaky_reLU(A), parameterised_reLU(A, 5), sep="\n\n", end="\n\n")
print(ELU(A), swish(A), softmax(A), sep="\n\n", end="\n\n")
```

```python
x = tf.placeholder(tf.float32)
sess = tf.Session()
print(sess.run(tf.nn.sigmoid(x), feed_dict = {x : A}), end="\n\n")
print(sess.run(tf.nn.tanh(x), feed_dict = {x : A}), end="\n\n")
print(sess.run(tf.nn.relu(x), feed_dict = {x : A}), end="\n\n")
print(sess.run(tf.nn.leaky_relu(x), feed_dict = {x : A}), end="\n\n")
print(sess.run(tf.nn.elu(x), feed_dict = {x : A}), end="\n\n")
print(sess.run(tf.nn.swish(x), feed_dict = {x : A}), end="\n\n")
print(sess.run(tf.nn.softmax(x), feed_dict = {x : A}), end="\n\n")
sess.close()
```
