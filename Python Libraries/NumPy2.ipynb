```python
#NumPy2
import numpy as np
```

```python
A=np.array([0, 0, 1, 1], dtype="bool")
B=np.array([0, 1, 0, 1], dtype="bool")
print(np.logical_not(A), end="\n\n")
print(np.logical_and(A, B), end="\n\n")
print(np.logical_or(A, B), end="\n\n")
print(np.logical_xor(A, B), end="\n\n")
```

```python
A=np.arange(1, 4, 0.5)
print(np.sin(A), np.cos(A), np.tan(A), sep="\n\n", end="\n\n")
print(np.log(A), np.log10(A), np.exp(A), sep="\n\n", end="\n\n")
print(np.square(A), np.sqrt(A), np.cbrt(A), np.power(A, 5), sep="\n\n", end="\n\n")
print(np.ceil(A), np.floor(A), np.round(A), sep="\n\n", end="\n\n")
```

```python
A=np.array([5, 3, 6])
B=np.array([2, 4, 1])
print(np.add(A, B), np.subtract(A, B), np.multiply(A, B), sep="\n\n", end="\n\n")
print(np.divide(A, B), np.mod(A, B), sep="\n\n", end="\n\n")
```

```python
A=np.array([2, 4, 1, 5, 3, 6])
print(np.sum(A), np.sort(A), sep="\n")
print(np.argsort(A), end="\n\n")
B=np.array([[2, 4, 1], [5, 3, 6]])
print(np.sum(B, axis=0), np.sort(B, axis=0), sep="\n")
print(np.argsort(B, axis=0), end="\n\n")
print(np.sum(B, axis=1), np.sort(B, axis=1), sep="\n")
print(np.argsort(B, axis=1), end="\n\n")
```

```python
print(np.argmax(A), np.max(A), sep="\n")
print(np.argmin(A), np.min(A), sep="\n", end="\n\n")
print(np.argmax(B, axis=0), np.max(B, axis=0), sep="\n")
print(np.argmin(B, axis=0), np.min(B, axis=0), sep="\n", end="\n\n")
print(np.argmax(B, axis=1), np.max(B, axis=1), sep="\n")
print(np.argmin(B, axis=1), np.min(B, axis=1), sep="\n", end="\n\n")
```

```python
print(np.mean(A), np.median(A), np.var(A), np.std(A), sep="\n", end="\n\n")
print(np.mean(B, axis=0), np.median(B, axis=0), np.var(B, axis=0), np.std(B, axis=0), sep="\n", end="\n\n")
print(np.mean(B, axis=1), np.median(B, axis=1), np.var(B, axis=1), np.std(B, axis=1), sep="\n", end="\n\n")
```

```python
print(np.insert(A, 1, 0), np.delete(A, 1), np.append(A, 0), sep="\n", end="\n\n")
print(np.insert(B, 1, [-1, 0, 1], axis=0), np.delete(B, 1, axis=0), np.append(B, [[-1, 0, 1]], axis=0), sep="\n", end="\n\n")
print(np.insert(B, 1, [-1, 1], axis=1), np.delete(B, 1, axis=1), np.append(B, [[-1], [1]], axis=1), sep="\n", end="\n\n")
```
