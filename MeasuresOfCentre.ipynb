```python
import random
import numpy as np
```

```python
P = []
for i in range(100) :
    P.append(random.randint(0, 999))
P = np.array(P)
print(P)
```

### Mean

```python
mean = np.mean(P)
print(mean)
```

### Median

```python
median = np.median(P)
print(median)
```

### Mode

```python
mode = median * 3 - mean * 2
print(mode)
```
