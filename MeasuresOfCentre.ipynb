```python
#MeasuresOfCentre
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("headbrain.csv")
df
```

```python
head_size=df["Head Size(cm^3)"].to_numpy()
print(head_size)
```

```python
#Mean
mean=np.mean(head_size)
print(mean)
```

```python
#Median
median=np.median(head_size)
print(median)
```

```python
#Mode
mode=(median*3)-(mean*2)
print(mode)
```
