```python
import tensorflow.compat.v1 as tf
import numpy as np
import math
tf.disable_eager_execution()
```

```python
def binary_step(x, n) :
  y=np.greater_equal(x, n)
  z=y.astype('int32')
  return z
```

```python
def linear(x, n) :
  y=x*n
  return y
```

```python
def sigmoid(x) :
  y=np.exp(-x)
  z=1/(1+y)
  return z
```

```python
def tanh(x) :
  y=np.tanh(x)
  return y
```

```python
def reLU(x) :
  y=[]
  for i in x :
    if i>=0 :
      y.append(i)
    else :
      y.append(0)
  return y
```

```python
def leaky_reLU(x) :
  y=[]
  for i in x :
    if i>=0 :
      y.append(i)
    else :
      y.append(i*0.2)
  return y
```

```python
def parameterised_reLU(x, n) :
  y=[]
  for i in x :
    if i>=0 :
      y.append(i)
    else :
      y.append(i*n)
  return y
```

```python
def ELU(x) :
  y=[]
  for i in x :
    if i>=0 :
      y.append(i)
    else :
      y.append(math.exp(i)-1)
  return y
```

```python
def swish(x) :
  y=np.exp(-x)
  z=x/(1+y)
  return z
```

```python
def softmax(x) :
  y=np.exp(x)
  z=y/np.sum(y)
  return z
```

```python
A=np.array([7, -8, 9, -4, 5, -6, 1, -2, 3, 0])
print(binary_step(A, 5), linear(A, 5), sigmoid(A), tanh(A), sep='\n\n', end='\n\n')
print(reLU(A), leaky_reLU(A), parameterised_reLU(A, 5), sep='\n\n', end='\n\n')
print(ELU(A), swish(A), softmax(A), sep='\n\n', end='\n\n')
```

```python
x=tf.placeholder(tf.float32)
sess=tf.Session()
print(sess.run(tf.nn.sigmoid(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.tanh(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.relu(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.leaky_relu(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.elu(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.swish(x), feed_dict={x:A}), end='\n\n')
print(sess.run(tf.nn.softmax(x), feed_dict={x:A}), end='\n\n')
sess.close()
```
