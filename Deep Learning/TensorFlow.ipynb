```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
x=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
```

```python
sess=tf.Session()
print(sess.run(tf.add(x, y), feed_dict={x:[4, 6], y:[2, 3]}), end='\n\n')
print(sess.run(tf.subtract(x, y), feed_dict={x:[4, 6], y:[2, 3]}), end='\n\n')
sess.close()
```

```python
sess=tf.InteractiveSession()
print(tf.multiply(x, y).eval(feed_dict={x:[4, 6], y:[2, 3]}), end='\n\n')
print(tf.divide(x, y).eval(feed_dict={x:[4, 6], y:[2, 3]}), end='\n\n')
print(tf.mod(x, y).eval(feed_dict={x:[4, 6], y:[2, 3]}), end='\n\n')
sess.close()
```

```python
x=tf.constant(5.5, tf.float32)
y=tf.constant(5.5, tf.float32)
```

```python
a=[tf.equal(x, y), tf.greater(x, y), tf.greater_equal(x, y), tf.less(x, y), tf.less_equal(x, y)]
b=[tf.sin(x), tf.cos(x), tf.tan(x), tf.log(x), tf.log1p(x), tf.exp(x)]
c=[tf.square(x), tf.sqrt(x), tf.pow(x, 5), tf.ceil(x), tf.floor(x), tf.round(x)]
```

```python
sess=tf.Session()
print(sess.run(a), end='\n\n')
print(sess.run(b), end='\n\n')
print(sess.run(c), end='\n\n')
sess.close()
```

```python
x=tf.Variable(tf.random_normal([2, 3]), tf.float32)
y=tf.Variable(tf.truncated_normal([2, 3]), tf.float32)
```

```python
a=[[1, 2]]
b=[[3],
   [4],
   [5]]
c=tf.matmul(a, b, transpose_a=True, transpose_b=True)
d=tf.cast(c, tf.float32)
e=tf.assign(x, d)
f=tf.assign(y, d)
```

```python
sess=tf.InteractiveSession()
e.eval()
f.eval()
print(x.eval(), end='\n\n')
print(y.eval(), end='\n\n')
sess.close()
```
