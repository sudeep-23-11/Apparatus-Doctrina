```python
import pandas as pd
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
print(df.info())
```

```python
sns.catplot(x='Survived', hue='Pclass', data=df, kind='count')
sns.catplot(x='Survived', hue='Sex', data=df, kind='count')
sns.catplot(x='SibSp', data=df, kind='count')
sns.catplot(x='Parch', data=df, kind='count')
sns.catplot(x='Survived', hue='Embarked', data=df, kind='count')
```

```python
sns.displot(x='Age', data=df, kind='hist')
sns.displot(x='Fare', data=df, kind='hist')
```

```python
print(df.isnull().sum())
```

```python
df.drop(['Age', 'Cabin'], axis=1, inplace=True)
df.fillna(value='S', inplace=True)
```

```python
Sex=LabelEncoder()
df['_Sex']=Sex.fit_transform(df['Sex'])
```

```python
Embarked=LabelEncoder()
df['_Embarked']=Embarked.fit_transform(df['Embarked'])
```

```python
df.drop(['Sex', 'Embarked'], axis=1, inplace=True)
```

```python
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
df.to_csv('titanic.csv', index=False)
```
