```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

```python
print(df.info())
```

```python
df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1, inplace=True)
```

```python
sns.catplot(x="Survived", hue="Pclass", data=df, kind="count")
sns.catplot(x="Survived", hue="Sex", data=df, kind="count")
sns.catplot(x="SibSp", data=df, kind="count")
sns.catplot(x="Parch", data=df, kind="count")
sns.catplot(x="Survived", hue="Embarked", data=df, kind="count")
plt.show()
```

```python
sns.displot(x="Age", data=df, kind="hist")
sns.displot(x="Fare", data=df, kind="hist")
plt.show()
```

```python
print(df.isnull().sum())
```

```python
df["Age"].fillna(value=df["Age"].mean(), inplace=True)
df["Fare"].fillna(value=df["Fare"].mean(), inplace=True)
```

```python
le = OneHotEncoder()
df_p = pd.DataFrame(le.fit_transform(df[["Pclass"]]).toarray())
df_p = df_p.rename(columns = {0 : "Pclass_1", 1 : "Pclass_2", 2 : "Pclass_3"})
df_s = pd.DataFrame(le.fit_transform(df[["Sex"]]).toarray())
df_s = df_s.rename(columns = {0 : "Sex_male", 1 : "Sex_female"})
df_e = pd.DataFrame(le.fit_transform(df[["Embarked"]]).toarray())
df_e = df_e.rename(columns = {0 : "Embarked_C", 1 : "Embarked_Q", 2 : "Embarked_S"})
df.drop(["Pclass", "Sex", "Embarked"], axis=1, inplace=True)
df = pd.concat([df, df_p, df_s, df_e], axis=1)
```

```python
X = df.drop(["Survived"], axis = 1)
Y = df["Survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
print(confusion_matrix(y_test, y_predict))
print(classification_report(y_test, y_predict))
```
