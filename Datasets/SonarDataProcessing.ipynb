```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv('sonar.csv')
df
```

```python
print(df.isnull().sum())
```

```python
MR=OneHotEncoder(handle_unknown='ignore')
_MR=pd.DataFrame(MR.fit_transform(df[['R']]).toarray())
_MR=_MR.rename(columns={0:'M', 1:'R'})
```

```python
df.drop(['R'], axis=1, inplace=True)
df=pd.concat([df, _MR], axis=1)
```

```python
df.to_csv('sonar.csv', index=False)
```
