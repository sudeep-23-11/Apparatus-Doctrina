```python
import pandas as pd
import numpy as np
```

```python
s=pd.Series(np.arange(1, 6, 1))
print(s)
```

```python
dict={'Name':['Sudeep', 'Aryan', 'Shivam', 'Jatin', 'Anshuman'], 'Rank':[1, 2, 3, 4, 5], 'PriceMoney':[1000.0, 500.0, 200.0, 100.0, 40.0]}
df=pd.DataFrame(dict)
df.index=['First', 'Second', 'Third', 'Fourth', 'Fifth']
df
```

```python
print(df.shape, df.dtypes, df.index, df.columns, df.T, sep='\n\n', end='\n\n')
print(df.head(3), df.tail(3), df.count(), df.value_counts(), sep='\n\n', end='\n\n')
print(df.max(), df.min(), df.mean(), df.median(), df.var(), df.std(), sep='\n\n', end='\n\n')
print(df.to_numpy(), df.describe(), sep='\n\n', end='\n\n')
print(df.info(), end='\n\n')
```

```python
df['Name'][2]='Yash'
print(df['Name'][2])
print(df['Name'])
df
```

```python
df1=df
df1.iloc[2, 0]='Shivam'
print(df1.iat[2, 0])
df
```

```python
df2=df.copy()
df2.loc['Third', 'Name']='Yash'
print(df2.at['Third', 'Name'])
df
```

```python
print(df.loc[['Second', 'Fourth'], ['Name', 'PriceMoney']], end='\n\n')
print(df.loc[:, ['Rank']], end='\n\n')
print(df.loc[['Third'], :], end='\n\n')
print(df.loc[(df['Rank']<5) & (df['PriceMoney']<1000)], end='\n\n')
print(df.loc[(df['Rank']==5) | (df['PriceMoney']==1000)], end='\n\n')
```

```python
print(df.sort_index(axis=0), end='\n\n')
print(df.sort_index(axis=0, ascending=False), end='\n\n')
print(df.sort_index(axis=1), end='\n\n')
print(df.sort_index(axis=1, ascending=False), end='\n\n')
print(df.sort_values(by='Name'), end='\n\n')
print(df.sort_values(by='Name', ascending=False), end='\n\n')
```
