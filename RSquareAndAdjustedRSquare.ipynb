```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('headbrain.csv')
df
```

```python
X=df['Head Size(cm^3)'].to_numpy()
Y=df['Brain Weight(grams)'].to_numpy()
```

```python
x_mean=np.mean(X)
y_mean=np.mean(Y)
N=len(df)
```

```python
#BestFitLine
n=0
d=0
for i in range(0, N, 1) :
  n+=((X[i]-x_mean)*(Y[i]-y_mean))
  d+=(X[i]-x_mean)**2
m=n/d
c=y_mean-(m*x_mean)
print(m, c)
```

```python
#RSquare
ss_r=0
ss_t=0
for i in range(0, N, 1) :
  ss_r+=(Y[i]-(m*X[i]+c))**2
  ss_t+=(Y[i]-y_mean)**2
r2=1-ss_r/ss_t
print(r2)
```

```python
#RSquare
ss_tr=0
ss_t=0
for i in range(0, N, 1) :
  ss_tr+=((m*X[i]+c)-y_mean)**2
  ss_t+=(Y[i]-y_mean)**2
r2=ss_tr/ss_t
print(r2)
```

```python
#AdjustedRSquare
p=1
ad_r2=1-((1-r2)*(N-1)/(N-p-1))
print(ad_r2)
```

```python
plt.plot(X, Y, 'ro')
plt.plot(X, (m*X+c), 'bo')
plt.show()
```
