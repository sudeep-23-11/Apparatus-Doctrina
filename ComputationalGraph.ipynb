```python
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()
```

```python
x=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
w=tf.Variable([0.5], tf.float32)
b=tf.Variable([-0.5], tf.float32)
y_predict=tf.add(tf.multiply(x, w), b)
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
print(sess.run(y_predict, feed_dict={x:[1, 2, 3, 4, 5]}))
```

```python
delta=tf.subtract(y, y_predict)
loss=tf.reduce_sum(tf.square(delta))
print(sess.run(loss, feed_dict={x:[1, 2, 3, 4, 5], y:[0, -1, -2, -3, -4]}))
```

```python
optimizer=tf.train.GradientDescentOptimizer(0.01)
train=optimizer.minimize(loss)
for epoch in range(1, 1001, 1) :
  sess.run(train, feed_dict={x:[1, 2, 3, 4, 5], y:[0, -1, -2, -3, -4]})
print(sess.run([w, b]))
sess.close()
```
