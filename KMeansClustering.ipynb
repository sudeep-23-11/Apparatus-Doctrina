```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('headbrain.csv')
df
```

```python
df.drop(['Gender', 'Age Range'], axis=1, inplace=True)
```

```python
X=df['Head Size(cm^3)'].to_numpy()
Y=df['Brain Weight(grams)'].to_numpy()
```

```python
X=np.reshape(X, (-1, 1))
Y=np.reshape(Y, (-1, 1))
```

```python
scale=MinMaxScaler()
X=scale.fit_transform(X)
Y=scale.fit_transform(Y)
```

```python
model=KMeans(n_clusters=5)
cluster=model.fit_predict(X, Y)
```

```python
print(model.cluster_centers_)
print(model.labels_)
```

```python
df['Cluster']=cluster
df1=df[df.Cluster==0]
df2=df[df.Cluster==1]
df3=df[df.Cluster==2]
df4=df[df.Cluster==3]
df5=df[df.Cluster==4]
```

```python
plt.scatter(df1['Head Size(cm^3)'], df1['Brain Weight(grams)'], color='r', label='Cluster 1')
plt.scatter(df2['Head Size(cm^3)'], df2['Brain Weight(grams)'], color='b', label='Cluster 2')
plt.scatter(df3['Head Size(cm^3)'], df3['Brain Weight(grams)'], color='g', label='Cluster 3')
plt.scatter(df4['Head Size(cm^3)'], df4['Brain Weight(grams)'], color='y', label='Cluster 4')
plt.scatter(df5['Head Size(cm^3)'], df5['Brain Weight(grams)'], color='k', label='Cluster 5')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.title('Head Size Vs Brain Weight')
plt.grid()
plt.legend()
plt.show()
```
